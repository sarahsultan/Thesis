### Mohephememizer:

```python
# read new lookup table
lookup_new = open("/content/mp_lookup.txt", "r")
lookup_new = lookup_new.readlines()

# remove "\n" from lookup_new
lookup_new1 = []
for i in lookup_new:
  lookup_new1.append(i.replace('\n', ''))
```

```python
#lookup table into dictinary
lookup_list = []
for i in lookup_new1:
  lookup_list.append(i.split(' ', 1))
# dictionary comprehension
#lookup_dict={x[0]:x[1] for x in lookup_list}  
```

```python
# separating tokens and their respective morphemes into lists
lookup_list1 = []
for m in lookup_list:
  for n in m:
    lookup_list1.append(n.split(' '))

# lookup dictionary
lookup_dict = iter(lookup_list1)
lookup_dict = {k: next(lookup_dict) for k, in lookup_dict}
```

```python
# read new vocabulary file
vocab_new = open("/content/mp_vocab.txt", "r")
vocab_new = vocab_new.readlines()

# remove "\n" from vocab_new
vocab_new1 = []
for i in vocab_new:
  vocab_new1.append(i.replace('\n', ''))
# vocab dictionary
vocab_dict = {}
for words in vocab_new1:
  vocab_dict[words] = [words]  
```

```python
# make one dictinary of two 
def mergeDictionary(dict_1, dict_2):
   dict_3 = {**dict_1, **dict_2}
   for key, value in dict_3.items():
       if key in dict_1 and key in dict_2:
               dict_3[key] = [value , dict_1[key]]
   return dict_3
dictionary = mergeDictionary(lookup_dict, vocab_dict)
```

```python
def whitespace_tokenizer(text):
  text = text.strip()
  if not text:
    return []
  tokens = text.split()
  return tokens

def tokenize(dic, text):
  output_tokens = []
  unknown_tokens = []
  for token in whitespace_tokenizer(text):
    if token in dic:
      output_tokens.append(dic[token])
    else:
      unknown_tokens.append(token)
  return output_tokens, unknown_tokens
```

```python
class Morphemetokenizer:
  def __init__(self, lookup_path, vocab_path, vocab, unk_token):
    self.lookup_path = lookup_path
    self.vocab_path = vocab_path
    self.vocab = vocab
    self.unk_token = unk_token

  # read lookup table
  lookup_new = open(self.lookup_path, "r")
  lookup_new = lookup_new.readlines()
  # remove "\n" from lookup_new
  lookup_new1 = []
  for i in lookup_new:
    lookup_new1.append(i.replace('\n', ''))
  #lookup table into dictinary
  lookup_list = []
  for i in lookup_new1:
    lookup_list.append(i.split(' ', 1))
  
  # separating tokens and their respective morphemes into lists
  lookup_list1 = []
  for m in lookup_list:
    for n in m:
      lookup_list1.append(n.split(' '))

  # lookup dictionary
  lookup_dict = iter(lookup_list1)
  lookup_dict = {k: next(lookup_dict) for k, in lookup_dict}
  # read new vocabulary file
  vocab_new = open(self.vocab, "r")
  vocab_new = vocab_new.readlines()

  # remove "\n" from vocab_new
  vocab_new1 = []
  for i in vocab_new:
    vocab_new1.append(i.replace('\n', ''))
  #vocab dictionary
  vocab_dict = {}
  for words in vocab_new1:
    vocab_dict[words] = [words]
  # make one dictinary of two 
  def mergeDictionary(dict_1, dict_2):
    dict_3 = {**dict_1, **dict_2}
    for key, value in dict_3.items():
      if key in dict_1 and key in dict_2:
        dict_3[key] = [value , dict_1[key]]
    return dict_3
  dictionary = mergeDictionary(lookup_dict, vocab_dict)
  
  
  def whitespace_tokenize(text):
    """Runs basic whitespace cleaning and splitting on a piece of text."""
    text = text.strip()
    if not text:
        return []
    tokens = text.split()
    return tokens

  def tokenize(self, text):
    output_tokens = []
    unknown_tokens = []
    for token in whitespace_tokenize(text):
      if token in dictionary:
        output_tokens.append(dictionary[token])
      else:
        unknown_tokens.append(token)
    return output_tokens, unknown_tokens
  
```
